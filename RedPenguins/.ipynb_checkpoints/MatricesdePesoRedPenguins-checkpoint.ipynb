```python
#funcion de activacion de valores pequeños 
#def f_act(X):
  #  activada = np.array([1/(1+math.exp(-x)) for x in X])
   # return activada
```

```python
def f_act(x):
    activada = []
    for i in x:
        try:
            r= 1/(1 + math.exp(-i))
        except OverflowError:
            r = 0
        activada.append(r)
    return activada
```

```python
#carga de datos 
penguins = sns.load_dataset('penguins')
penguins
```

```python
xcols = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
especie = penguins['species'].unique()
especie
```

```python
X = penguins[xcols].copy()
X.insert(0,'bias',1)
X
```

```python
#Prediccion 
prediccion = []
for index, fila in X.iterrows():
    capa1 = f_act(fila.dot(W1))
    capa1 = np.insert(capa1,0,1)
    capa2 = f_act(capa1.dot(W2))
    capa2 = np.insert(capa2,0,1)
    capa3 = f_act(capa2.dot(W3))
    capa3 = np.insert(capa3,0,1)
    salida = f_act(capa3.dot(W4))
    prediccion.append(especie[np.argmax(salida)])
```

```python
prediccion
```

```python
penguins['Prediccion'] = prediccion 
penguins.head()
```
