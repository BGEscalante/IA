```python
import math 
import numpy as np
import seaborn as sns 
```

```python
#matriz de pesos 1ra. capa
W1= np.array([
              [-3.69211462961044,-42.6152693749059],
              [-0.609243431773111,-0.173515799905688],
              [-1.08038855081607,-26.1856656230522],
              [0.803576313518212,16.8291714008646],
              [3.4190210289766,59.2539591591492]
]) 
```

```python
#matriz de pesos 2da. capa
W2= np.array([
            [-1.66026224716512,23.9692039655777,0.574110982582267],
            [-0.0884830512482629,-267.924604402339,-1.98998360908763],
            [3.05850518781859,68.5808229144055,-5.08815570750833]
]) 
```

```python
#matriz de pesos 3ra. capa
W3= np.array([
            [0.314049863670979,0.203500197039452,0.264034861658445],
            [-0.40096831535927,-0.250841249799398,0.926166813777208],
            [-0.00365427249201641,1.01464427671184,-1.01468131554472],
            [1.17833983734437,-1.84195458230484,0.942360160831213]

]) 
```

```python
#funcion de activacion de valores pequeños 
def f_act(X):
    activada = np.array([1/(1+math.exp(-x)) for x in X])
    return activada
```

```python
#carga de datos 
iris = sns.load_dataset('iris')
iris
```

```python
xcols = ['sepal_length', 'sepal_width','petal_length','petal_width']
especie = iris['species'].unique()
especie
```

```python
X = iris[xcols].copy()
X.insert(0,'bias',1)
X
```

```python
#Prediccion 
prediccion = []
for index, fila in X.iterrows():
    capa1 = f_act(fila.dot(W1))
    capa1 = np.insert(capa1,0,1)
    capa2 = f_act(capa1.dot(W2))
    capa2 = np.insert(capa2,0,1)
    salida = f_act(capa2.dot(W3))
    prediccion.append(especie[np.argmax(salida)])
```

```python
prediccion
```

```python
iris['Prediccion'] = prediccion 
iris.head()
```
