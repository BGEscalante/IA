```python
##PREPROCESAMIENTO 

#cargar  modulos necesarios 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 

```

```python
#cargar el conjunto de datos 
data = sns.load_dataset('iris')
data.head(5)
```

```python
##resumen de datos 
data.describe()
```

```python
##ver cual es la distribucion de cada clase
data.groupby('species').size()
```

```python
#carga de bibliotecas especificas para modelos de inteligencia artificial
from sklearn.model_selection import train_test_split
```

```python
#separacion de datos en grupo de entrenamiento y prueba
train, test = train_test_split(data,test_size=0.4,stratify=data['species'],random_state=19)
train.head(10)
```

```python
#analisis exploratorio 
#para evaluar las variables 
#histogramas para observar la distribucion de los 4 atributos 
n_bins = 10
fig, axs = plt.subplots(2,2) #arreglo de 2x2
axs[0,0].hist(train['sepal_length'],bins=n_bins)
axs[0,0].set_title('longitud de sepalo')
axs[0,1].hist(train['sepal_width'],bins=n_bins)
axs[0,1].set_title('ancho de sepalo')
axs[1,0].hist(train['petal_length'],bins=n_bins)
axs[1,0].set_title('longitud de petalo')
axs[1,1].hist(train['petal_width'],bins=n_bins)
axs[1,1].set_title('ancho de petalo')
fig.tight_layout(pad=1.0);
```

```python
#diagrmas boxplot
fn = ['sepal_length','sepal_width','petal_length','petal_width']
cn = ['setosa','versicolor','virginica']
fig, axs = plt.subplots(2,2)
sns.boxplot(x='species',y=fn[0],data=train,order=cn,ax=axs[0,0])
sns.boxplot(x='species',y=fn[1],data=train,order=cn,ax=axs[0,1])
sns.boxplot(x='species',y=fn[2],data=train,order=cn,ax=axs[1,0])
sns.boxplot(x='species',y=fn[3],data=train,order=cn,ax=axs[1,1])
fig.tight_layout(pad=1.0);
```

```python
#diagrama de violin 
sns.violinplot(x='species',y=fn[2],data=train,size=5, order=cn, palette ='colorblind')
```

```python
#diagrama de dispersion de atributos emparejados 
sns.pairplot(train,hue='species', height=2, palette = 'pastel')
```

```python
##matrices de correlacion 
corrmat = train.corr(numeric_only = True)
corrmat 
```

```python
sns.heatmap(corrmat, annot =True,square=True)
```

```python
##coordenadas paralelas 
pd.plotting.parallel_coordinates(train,'species',color=['blue','red','green']);
```

```python
#arbol de decision (clasificacion)
#separacion de la etiqueta de clase y atributos 
```

```python
#Separacion objetivo - explicativas de cada grupo 
#grupo de entrenamiento 
X_train = train[['sepal_length','sepal_width', 'petal_length', 'petal_width']]
Y_train = train['species']
#grupo de pruebas
X_test = test[['sepal_length','sepal_width', 'petal_length', 'petal_width']]
Y_test = test['species']
print(X_train.head(5))
print(Y_train.head(5))
```

```python
#cargar bibliotecas especificas
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
#creacion del modelo del arbol de decision
mod_dt = DecisionTreeClassifier(max_depth=3,random_state=1)
mod_dt.fit(X_train,Y_train)
```

```python
#visualizacion 
plt.figure(figsize=(10,8))
plot_tree(mod_dt, feature_names=list(mod_dt.feature_names_in_),
          class_names=cn, filled=True);
```

```python
#verificacion del modelo 
##Predecir resultados utilizando el modelo
predicccion = mod_dt.predict(X_test)
```

```python
##modulo para metricas de eficiencia de modelo 
from sklearn import metrics
```

```python
print('la precision del arbol de decision es:{:.3f}',format(metrics.accuracy_score(predicccion,Y_test))) 
```
